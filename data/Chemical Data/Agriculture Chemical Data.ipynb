```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
corn_fert = pd.read_csv("./corn_fert.csv")
potato_fert = pd.read_csv("./potato_fert.csv")
corn_pesti = pd.read_csv("./corn_pesti.csv")
```

```python
corn_fert = corn_fert[['State','Year','CORN - APPLICATIONS, MEASURED IN LB / ACRE / YEAR, AVG  -  <b>VALUE</b>', 
                      'CORN - APPLICATIONS, MEASURED IN NUMBER, AVG  -  <b>VALUE</b>',
                      'CORN - TREATED, MEASURED IN PCT OF AREA PLANTED, AVG  -  <b>VALUE</b>']]
corn_fert.colums = ['State', 'Year', 'Fert per year', 'Fert per application', 'Fert per area planted']

corn_pesti = corn_pesti[['State','Year','CORN - APPLICATIONS, MEASURED IN LB  -  <b>VALUE</b>', 
                      'CORN - APPLICATIONS, MEASURED IN NUMBER, AVG  -  <b>VALUE</b>',
                      'CORN - TREATED, MEASURED IN PCT OF AREA PLANTED, AVG  -  <b>VALUE</b>']]
```

```python
print(corn_fert)
print(corn_pesti)
```

```python
ax = plt.axes(projection='3d')
h = plt.plot(corn_fert['Year'], corn_fert['CORN - APPLICATIONS, MEASURED IN LB / ACRE / YEAR, AVG  -  <b>VALUE</b>'], corn_fert['CORN - TREATED, MEASURED IN PCT OF AREA PLANTED, AVG  -  <b>VALUE</b>'], '.b', markersize=8, label='Non Pareto-optimal')
ax.set_xlabel('Year')
ax.set_ylabel('Fertilizer per year')
ax.set_zlabel('Fertilizer per area')
_ = plt.title('Corn Fertilizer', fontsize=15)
plt.show()
```

```python
ax = plt.axes(projection='3d')
h = plt.plot(corn_pesti['Year'], corn_pesti['CORN - APPLICATIONS, MEASURED IN LB  -  <b>VALUE</b>'], corn_pesti['CORN - TREATED, MEASURED IN PCT OF AREA PLANTED, AVG  -  <b>VALUE</b>'], '.b', markersize=8, label='Non Pareto-optimal')
ax.set_xlabel('Year')
ax.set_ylabel('Pesticide per year')
ax.set_zlabel('Pesticide per area')
_ = plt.title('Corn Fertilizer', fontsize=15)
plt.show()
```
